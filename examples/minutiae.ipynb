# End-to-end minutiae extraction

```python
import cv2 as cv
import pyfing as pf
from pyfing.utils.notebook import show
from pyfing.utils.drawing import draw_minutiae
```

```python
# Paths
PATH_FVC = '../../datasets/'
```

Example: end-to-end minutiae extraction using the LEADER method.

```python
fingerprint = cv.imread(PATH_FVC + 'fvc2000/db2_b/101_1.png', cv.IMREAD_GRAYSCALE)
alg = pf.Leader()
minutiae = alg.run(fingerprint)
show(fingerprint, draw_minutiae(fingerprint, minutiae))
```
